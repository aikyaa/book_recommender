```python
from statistics import correlation

import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd
```

```python
books=pd.read_csv(f"{path}/books.csv")
books
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
missing_stats=plt.axes()
sns.heatmap(books.isna().transpose(), cbar=False, ax=missing_stats)
plt.xlabel("Fields")
plt.ylabel("Missing Values")
plt.show()
```

```python
import numpy as np

books["missing_description"]=np.where(books["description"].isna(),1,0)
books["age"]=2025-books["published_year"]
```

```python
columns_of_interest=["num_pages", "age", "missing_description", "average_rating"]
correlation_matrix= books[columns_of_interest].corr(method="spearman")

sns.set_theme(style="whitegrid")
plt.figure(figsize=(8,6))
heatmap=sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", cbar_kws={"label": "Spearman Correlation"})
heatmap.set_title("Correlation Heatmap")
plt.show()

#strong correlations: +/- 1
```

```python
books_complete_details = books[~(books["description"].isna()) &
        ~(books["num_pages"].isna()) &
        ~(books["published_year"].isna()) &
        ~(books["average_rating"].isna())
]
#we can discard them since the percentage is negligible and won't make our recommender biased

books_complete_details
```

```python
books_complete_details["categories"].value_counts().reset_index().sort_values("count", ascending=False)
```

```python
books_complete_details["words_in_description"]= books_complete_details["description"].str.split().str.len()
#gives word count
books_complete_details
```

```python
#define meaningful descriptions by setting a cutoff
books_complete_details.loc[books_complete_details["words_in_description"].between(1,4), "description"]
```

```python
books_complete_details.loc[books_complete_details["words_in_description"].between(5,14), "description"]
```

```python
books_complete_details.loc[books_complete_details["words_in_description"].between(15,24), "description"]
```

```python
books_complete_details.loc[books_complete_details["words_in_description"].between(25,34), "description"]
```

```python
books_valid_description= books_complete_details[books_complete_details["words_in_description"]>=25]
books_valid_description
```

```python
books_valid_description["title_and_subtitle"]=(np.where(
    books_valid_description["subtitle"].isna(), books_valid_description["title"],
    books_valid_description[["title", "subtitle"]].astype(str).agg(":".join, axis=1)
))
books_valid_description
```

```python
books_valid_description["tagged_description"]=books_valid_description[["isbn13", "description"]].astype(str).agg(" ".join, axis=1)
books_valid_description
```

```python
(
    books_valid_description
    .drop(["subtitle","missing_description", "words_in_description", "age"], axis=1)
    .to_csv("books_cleaned.csv", index=False)
)
```
